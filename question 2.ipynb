```python
import numpy as np
import pandas as ps
import matplotlib.pyplot as plt
import seaborn as sns
laa = ps.read_csv('/content/diabetes.csv')


np.random.seed(123)

sample = laa.sample(n=25)

sample_percentile_bmi = np.percentile(sample['BMI'], 98)


population_percentile_bmi = np.percentile(laa['BMI'], 98)


labels = ['Population 98th Percentile BMI', 'Sample 98th Percentile BMI']
values = [population_percentile_bmi, sample_percentile_bmi]

plt.bar(labels, values, color=['pink', 'blue'])
plt.title('Comparison of 98th Percentile BMI')
plt.ylabel('BMI Value')
plt.show()


print(f"Population of 98th Percentile BMI: {population_percentile_bmi}")
print(f"Sample of 98th Percentile BMI: {sample_percentile_bmi}")
```
