```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plot

laa = pd.read_csv('/content/diabetes.csv')


np.random.seed(810)

sample = laa.sample(n=25)
sample_mean_glucose = sample['Glucose'].mean()
sample_highest_glucose = sample['Glucose'].max()

population_mean_glucose = laa['Glucose'].mean()
population_highest_glucose = laa['Glucose'].max()

labels = ['Population Mean', 'Population Highest', 'Sample Mean', 'Sample Highest']
values = [population_mean_glucose, population_highest_glucose, sample_mean_glucose, sample_highest_glucose]

plot.bar(labels, values, color=['blue', 'blue', 'pink', 'pink'])
plot.title('Comparison of Glucose Statistics')
plot.ylabel('Glucose Value')
plot.show()
```
